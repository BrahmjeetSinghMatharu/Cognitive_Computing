```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
# Generate x values from -10 to 10 using np.linspace().
# Compute y values for:
# y= M⋅x2
# y= M⋅sin(x)
# Plot both functions in a single figure:
# Use different colors and line styles.
# Add legend, grid, and title.

M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'r-', label=f'y = {M}x^2')
plt.plot(x, y2, 'b--', label=f'y = {M}sin(x)')
plt.legend()
plt.grid()
plt.title("Mathematical Function Plots")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Q.2 Create a dataset of five subjects and scores.
# Convert it into a Pandas DataFrame.
# Plot the scores using a Seaborn bar plot with:
# Different colors for each bar.
# Annotations on top of each bar.
# Title, axis labels, and grid.


data = {'Subjects': ['Math', 'Science', 'History', 'English', 'Art'],
        'Scores': [88, 92, 75, 85, 90]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.barplot(x='Subjects', y='Scores', hue='Subjects', data=df, palette='pastel', legend=False)

for index, value in enumerate(df['Scores']):
    plt.text(index, value + 1, str(value), ha='center', fontsize=12)

plt.title("Student Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid()
plt.show()
```

```python
# Q3. Write a Python script to select your roll number as seed for NumPy and
# then generate a dataset of 50 values using np.random.randn().
# Create a 2x2 subplot layout displaying:
# Line plot showing cumulative sum.
# Scatter plot with random noise.
# Customize with titles, axis labels, and grids.

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)
cumsum = np.cumsum(data)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(cumsum, 'g-', label="Cumulative Sum")
axes[0].set_title("Cumulative Sum Plot")
axes[0].legend()
axes[0].grid()

axes[1].scatter(range(50), data, c='red', label="Random Noise")
axes[1].set_title("Scatter Plot")
axes[1].legend()
axes[1].grid()

plt.tight_layout()
plt.show()
```

```python
# Q.4 Download Data-set from the below link
# https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
# Apply ‘seaborn’ library to do the following.
# 1. Read Total profit of all months and show it using a line plot.
# 2. Read all product sales data and show it using a multiline plot.
# 3. Plot bar chart for all the features/attributes.


df_sales = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")


plt.figure(figsize=(8, 6))
plt.plot(df_sales['month_number'], df_sales['total_profit'], marker='o', linestyle='-', color='b')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
for column in df_sales.columns[1:-1]:
    plt.plot(df_sales['month_number'], df_sales[column], marker='o', label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
df_sales.set_index('month_number').plot(kind='bar', figsize=(10, 6))
plt.title("Company Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.grid()
plt.show()
```
